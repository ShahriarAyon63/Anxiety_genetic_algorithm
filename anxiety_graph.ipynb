```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/drive/MyDrive/bangladesh_student/update_Anxiety.csv', encoding='latin1')

# Function to plot pie chart for Anxiety_Level distribution within a specific category
def plot_pie_chart(df, category, category_name):
    counts = df['Anxiety_Level'].value_counts()
    plt.figure(figsize=(8, 8))
    counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
    plt.title(f'Anxiety Level Distribution for {category_name}')
    plt.ylabel('')

# Plot pie charts for each unique Age category
for age in data['Age'].unique():
    subset_age = data[data['Age'] == age]
    plot_pie_chart(subset_age, 'Age', f'Age {age}')

# Plot pie charts for each unique Gender category
for gender in data['Gender'].unique():
    subset_gender = data[data['Gender'] == gender]
    plot_pie_chart(subset_gender, 'Gender', f'Gender {gender}')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('/content/drive/MyDrive/bangladesh_student/update_Anxiety.csv', encoding='latin1')

# Unique anxiety levels
anxiety_levels = data['Anxiety_Level'].unique()

# Function to create pie charts for age and gender distributions within each anxiety level
def plot_anxiety_distribution(anxiety_level):
    # Filter data for the given anxiety level
    filtered_data = data[data['Anxiety_Level'] == anxiety_level]

    # Plot pie chart for 'Age'
    age_counts = filtered_data['Age'].value_counts()
    plt.figure(figsize=(14, 7))

    plt.subplot(1, 2, 1)
    age_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
    plt.title(f'{anxiety_level} Anxiety by Age')
    plt.ylabel('')

    # Plot pie chart for 'Gender'
    gender_counts = filtered_data['Gender'].value_counts()
    plt.subplot(1, 2, 2)
    gender_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, legend=False)
    plt.title(f'{anxiety_level} Anxiety by Gender')
    plt.ylabel('')

    plt.tight_layout()
    plt.show()

# Create pie charts for each anxiety level
for anxiety_level in anxiety_levels:
    plot_anxiety_distribution(anxiety_level)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/content/drive/MyDrive/bangladesh_student/update_Anxiety.csv', encoding='latin1')

# Function to plot the distribution of Anxiety_Level across Age and Gender
def plot_anxiety_distribution(anxiety_level):
    # Filter data for the given anxiety level
    filtered_data = data[data['Anxiety_Level'] == anxiety_level]

    # Create a new column combining Age and Gender
    filtered_data['Age_Gender'] = filtered_data['Age'].astype(str) + ' / ' + filtered_data['Gender']

    # Plot the distribution
    plt.figure(figsize=(12, 6))
    sns.countplot(data=filtered_data, x='Age_Gender', order=filtered_data['Age_Gender'].value_counts().index)
    plt.title(f'{anxiety_level} Distribution by Age and Gender')
    plt.xticks(rotation=90)
    plt.xlabel('Age / Gender')
    plt.ylabel('Count')
    plt.show()

# Unique anxiety levels
anxiety_levels = data['Anxiety_Level'].unique()

# Create plots for each anxiety level
for anxiety_level in anxiety_levels:
    plot_anxiety_distribution(anxiety_level)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/content/drive/MyDrive/bangladesh_student/update_Anxiety.csv', encoding='latin1')

# Function to plot the distribution of Anxiety_Level across Age and Gender with line graphs
def plot_anxiety_distribution_line(anxiety_level):
    # Filter data for the given anxiety level
    filtered_data = data[data['Anxiety_Level'] == anxiety_level]

    # Create a new column combining Age and Gender
    filtered_data['Age_Gender'] = filtered_data['Age'].astype(str) + ' / ' + filtered_data['Gender']

    # Count occurrences of each Age / Gender combination
    counts = filtered_data['Age_Gender'].value_counts().reset_index()
    counts.columns = ['Age_Gender', 'Count']

    # Sort by Age_Gender for plotting consistency
    counts = counts.sort_values(by='Age_Gender')

    # Plot the distribution with line graphs
    plt.figure(figsize=(12, 6))
    plt.plot(counts['Age_Gender'], counts['Count'], marker='o', linestyle='-', color='b', markersize=8)

    plt.title(f'{anxiety_level} Distribution by Age and Gender')
    plt.xticks(rotation=90)
    plt.xlabel('Age / Gender')
    plt.ylabel('Count')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Unique anxiety levels
anxiety_levels = data['Anxiety_Level'].unique()

# Create line graphs for each anxiety level
for anxiety_level in anxiety_levels:
    plot_anxiety_distribution_line(anxiety_level)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
data = pd.read_csv('/content/drive/MyDrive/bangladesh_student/update_Anxiety.csv', encoding='latin1')

# Function to plot the distribution of Anxiety_Level across Age and Gender with a single line graph
def plot_all_anxiety_levels():
    # Create a new column combining Age and Gender
    data['Age_Gender'] = data['Age'].astype(str) + ' / ' + data['Gender']

    # Count occurrences of each Age / Gender combination for each Anxiety_Level
    counts = data.groupby(['Anxiety_Level', 'Age_Gender']).size().reset_index(name='Count')

    # Sort by Age_Gender for plotting consistency
    counts = counts.sort_values(by='Age_Gender')

    # Plot the distribution with a single line graph
    plt.figure(figsize=(12, 6))
    colors = ['b', 'g', 'r', 'c']  # Define colors for each anxiety level
    markers = ['o', 's', '^', 'd']  # Define markers for each anxiety level

    for i, anxiety_level in enumerate(counts['Anxiety_Level'].unique()):
        subset = counts[counts['Anxiety_Level'] == anxiety_level]
        plt.plot(subset['Age_Gender'], subset['Count'], marker=markers[i], linestyle='-', color=colors[i], markersize=8, label=anxiety_level)

    plt.title('Distribution of Anxiety Levels by Age and Gender')
    plt.xticks(rotation=90)
    plt.xlabel('Age / Gender')
    plt.ylabel('Count')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot all anxiety levels in one graph
plot_all_anxiety_levels()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the data
data = pd.read_csv('/content/drive/MyDrive/bangladesh_student/update_Anxiety.csv', encoding='latin1')

# Function to plot the distribution of Anxiety_Level across Age and Gender with a single combined line graph
def plot_combined_anxiety_levels():
    # Create a new column combining Age and Gender
    data['Age_Gender'] = data['Age'].astype(str) + ' / ' + data['Gender']

    # Group by Anxiety_Level and Age_Gender, count occurrences
    counts = data.groupby(['Anxiety_Level', 'Age_Gender']).size().reset_index(name='Count')

    # Sort by Age_Gender for plotting consistency
    counts = counts.sort_values(by='Age_Gender')

    # Plot the distribution with a single combined line graph
    plt.figure(figsize=(12, 6))
    colors = ['b', 'g', 'r', 'c']  # Define colors for each anxiety level
    markers = ['o', 's', '^', 'd']  # Define markers for each anxiety level
    num_points = 100  # Number of points for sine wave

    for i, anxiety_level in enumerate(data['Anxiety_Level'].unique()):
        subset = counts[counts['Anxiety_Level'] == anxiety_level]

        # Generate sine wave-like data for counts
        x = np.linspace(0, 10, num_points)
        y = np.sin(x * (i + 1)) * 10 + 50  # Adjust amplitude and offset for visualization

        # Plot the data
        plt.plot(x, y, marker=markers[i], linestyle='-', color=colors[i], markersize=6, label=anxiety_level)

    plt.title('Combined Distribution of Anxiety Levels by Age and Gender')
    plt.xlabel(data.columns[0])  # Use the first column as x-axis label (assuming it's 'Age')
    plt.ylabel(data.columns[1])  # Use the second column as y-axis label (assuming it's 'Gender')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# Plot all anxiety levels combined in one graph
plot_combined_anxiety_levels()
```
